```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
path_pickles = "../Models/"

pickle_list =["df_models_gnbc.pickle",
             "df_models_dtc.pickle",
             "df_models_lrc.pickle",
             "df_models_svm.pickle",
             "df_models_rfc.pickle",
              "df_models_knn.pickle",
             "df_models_gbc.pickle",
             "df_models_xgbc.pickle"]
```

```python
df_summary=pd.DataFrame()
```

```python
for pickles in pickle_list:
    path = path_pickles + pickles
    with open(path,'rb') as data:
        df = pickle.load(data)
    df_summary=df_summary.append(df)
df_summary=df_summary.reset_index().drop('index',axis=1)
```

```python
df_summary
```

```python
df_summary.sort_values('Test Set Accuracy',ascending=True)
```

```python
d = df_summary.copy()
short_name=['GNBC',"KNN","DTC","SVM","LRC","RFC","GBC","EGBC"]
d['Model_Name']=short_name
d
```

```python
#Plotting the accuracy plot
d.plot(x="Model_Name", y=["Training Set Accuracy", "Test Set Accuracy"], kind="bar",figsize=(9,8))
plt.title('Accuracy PLot of Different Models', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
plt.xlabel('Models', fontsize=16)
plt.savefig("../Images/accuracy.png")
```
