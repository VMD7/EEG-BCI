```python
#Plotting the fft spectrum of a set of data
sample_a = df.loc[0, 'fft_0_a':'fft_749_a']
plt.figure(figsize=(16, 10))
plt.plot(range(len(sample_a)), sample_a)
plt.title("Features fft_0_a through fft_749_a")
plt.savefig("../Images/Features fft_0_a through fft_749_a.png")
```

```python
#Plotting the fft spectrum of b set of data
sample_b = df.loc[0, 'fft_0_b':'fft_749_b']
plt.figure(figsize=(16, 10))
plt.plot(range(len(sample_b)), sample_b)
plt.title("Features fft_0_b through fft_749_b")
plt.savefig("../Images/Features fft_0_b through fft_749_b.png")
```

```python
#Plotting the count of labels in the dataset
plt.figure(figsize=(12,5))
sns.countplot(x=df.label)
plt.title('Class Distribution of Emotion Sentiment', fontsize=16)
plt.ylabel('Class Counts', fontsize=16)
plt.xlabel('Class Label', fontsize=16)
plt.xticks(rotation='vertical');
plt.savefig("../Images/ClassDistributionOfImages.png")
```
